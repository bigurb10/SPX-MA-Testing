```python
%matplotlib inline
import pandas as pd
simple_returns_dict = dict()
exponential_returns_dict = dict()
output_to_csv1 = []
output_to_csv2 = []
#running the simple and exponential moving average tests
for i in range(2, 301):
    df = pd.read_csv('SPX_Data.csv')
    df.columns = df.columns.str.strip()
    df['Pct Change'] = df['Adj Close'].pct_change()
    df['Simple Moving Average'] = df['Adj Close'].rolling(window=i).mean().round(5)
    df['Exponential Moving Average'] = df['Adj Close'].ewm(span=i,adjust=False).mean().round(5)
    df['Criteria'] = df['Adj Close'] >= df['Simple Moving Average']
    df['Criteria 2'] = df['Adj Close'] >= df['Exponential Moving Average']
    #actual model
    df['Buy and Hold'] = 100*(1+df['Pct Change']).cumprod()
    df['Simple Model'] = 100*(1+df['Criteria'].shift(1)*df['Pct Change']).cumprod()
    df['Exponential Model'] = 100*(1+df['Criteria 2'].shift(1)*df['Pct Change']).cumprod()
    #simple moving average model result
    simple_start_portfolio = df['Simple Model'].iloc[300]
    simple_end_portfolio = df['Simple Model'].iloc[-1]
    years = (df['Simple Model'].count()+1-300)/252
    simple_average_return = (simple_end_portfolio/simple_start_portfolio)**(1/years)-1
    simple_returns_dict[i] = simple_average_return
    print('This SIMPLE model yields an average of', simple_average_return, 'per year', 'using', i, 'days')
    output_to_csv1.append(('The ' + str(i) + ' sma',simple_average_return ))
    #exponential moving average model result
    exponential_start_portfolio = df['Exponential Model'].iloc[300]
    exponential_end_portfolio = df['Exponential Model'].iloc[-1]
    exponential_average_return = (exponential_end_portfolio/exponential_start_portfolio)**(1/years)-1
    exponential_returns_dict[i] = exponential_average_return
    print('This EXPONENTIAL model yields an average of', exponential_average_return, 'per year', 'using', i, 'days')
    output_to_csv2.append(('The ' + str(i) + ' ema',exponential_average_return ))
    #buy and hold the market result
    start_spx = df['Adj Close'].iloc[300]
    end_spx = df['Adj Close'].iloc[-1]
    spx_average_return = (end_spx/start_spx)**(1/years)-1
print('Buy and hold average return is', spx_average_return, 'per year')
out_df1 = pd.DataFrame(output_to_csv1, columns=['SMA', 'average_return'])
out_df2 = pd.DataFrame(output_to_csv2, columns=['EMA', 'average_return'])
#find best simple moving average
max_simple_return = 0
max_simple_days = 0
for key in simple_returns_dict:
    if simple_returns_dict[key] > max_simple_return:
        max_simple_return = simple_returns_dict[key]
        max_simple_days = key
    else:
        max_simple_return = max_simple_return
        max_simple_days = max_simple_days
print('The best SIMPLE MOVING AVERAGE return is', max_simple_return, 'which uses', max_simple_days, 'simple moving average days')
#find best exponential moving average
max_exponential_return = 0
max_exponential_days = 0
for key in exponential_returns_dict:
    if exponential_returns_dict[key] > max_exponential_return:
        max_exponential_return = exponential_returns_dict[key]
        max_exponential_days = key
    else:
        max_exponential_return = max_exponential_return
        max_exponential_days = max_exponential_days
print('The best EXPONENTIAL MOVING AVERAGE return is', max_exponential_return, 'which uses', max_exponential_days, 'exponential moving average days')
#print(df)

out_df1.to_csv('movingaveragetest1.csv')
out_df2.to_csv('movingaveragetest2.csv')
```
